```python
import pandas as pd
```

Буду делать через пандас

```python
df = pd.read_json("coffee_finetuning_20240914_witi_total_ru.jsonl", lines=True)
df.shape
```

Исходный размер

```python
df = df.drop_duplicates(
    subset=["instruction", "response"]
)
print("Размер после удаления дубликатов:", df.shape)
```

```python
df = df.dropna(
    subset=["instruction", "response"]
)
print("Размер после удаления строк с пропущенными значениями:", df.shape)
```

Удаление пустых строк

```python
df = df[
    (df["instruction"] != "") &
    (df["response"] != "")
]
print("Размер после удаления строк с пустыми значениями:", df.shape)
```

Удаление коротких ответов

```python
df = df[
    df["response"].str.len() >= 10
]
print("Размер после удаления строк с короткими ответами:", df.shape)
```

Чтобы оставиить только 250 строк я сначала перемещал весь датасет чтобы вопросы не были похожими

```python
df = df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

df = df.head(250)
```

Потом сохранил это все

```python
df.to_json(
    "cleaned_dataset.jsonl",
    orient="records",
    lines=True,
    force_ascii=False
)
```
